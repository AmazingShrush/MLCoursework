```python
from sklearn.model_selection import train_test_split

# assuming X is your data and y is your target variable
X_train, X_test, y_train, y_test = train_test_split(heart_data, heart_target, test_size=0.2, random_state=42)
```

```python
from sklearn.impute import SimpleImputer

imputer=SimpleImputer(strategy="median")
imputer.fit(X_train)
imputer.statistics_
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

preprocessing_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std_scaler',StandardScaler())
])

X_train_prepared = preprocessing_pipeline.fit_transform(X_train)
```

```python
X_train_prepared
```

```python
from sklearn.linear_model import LogisticRegression

# Train logistic regression model without regularisation
logistic_regression = LogisticRegression(max_iter=1000)
logistic_regression.fit(X_train_prepared, np.ravel(y_train))

# Train logistic regression model with L1 regularisation
l1_logistic_regression = LogisticRegression(penalty='l1', solver='liblinear', max_iter=1000)
l1_logistic_regression.fit(X_train_prepared,np.ravel(y_train))

# Train logistic regression model with L2 regularisation
l2_logistic_regression = LogisticRegression(penalty='l2', max_iter=1000)
l2_logistic_regression.fit(X_train_prepared,np.ravel(y_train))
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Define logistic regression models with L1 and L2 regularization
# logreg_l1 = LogisticRegression(penalty='l1', solver='liblinear')
# logreg_l2 = LogisticRegression(penalty='l2')

# Perform cross-validation
cv_scores_l1 = cross_validate(l1_logistic_regression, X_train_prepared, np.ravel(y_train), cv=5,
                              scoring=['accuracy', 'precision', 'recall', 'f1'])
cv_scores_l2 = cross_validate(l2_logistic_regression, X_train_prepared, np.ravel(y_train), cv=5,
                              scoring=['accuracy', 'precision', 'recall', 'f1'])

# Extract the mean scores for each metric
accuracy_l1 = cv_scores_l1['test_accuracy'].mean()
precision_l1 = cv_scores_l1['test_precision'].mean()
recall_l1 = cv_scores_l1['test_recall'].mean()
f1_l1 = cv_scores_l1['test_f1'].mean()

accuracy_l2 = cv_scores_l2['test_accuracy'].mean()
precision_l2 = cv_scores_l2['test_precision'].mean()
recall_l2 = cv_scores_l2['test_recall'].mean()
f1_l2 = cv_scores_l2['test_f1'].mean()

# Print the results
print('Logistic Regression with L1 Regularization')
print(f'Accuracy: {accuracy_l1:.4f}')
print(f'Precision: {precision_l1:.4f}')
print(f'Recall: {recall_l1:.4f}')
print(f'F1 score: {f1_l1:.4f}')

print('\nLogistic Regression with L2 Regularization')
print(f'Accuracy: {accuracy_l2:.4f}')
print(f'Precision: {precision_l2:.4f}')
print(f'Recall: {recall_l2:.4f}')
print(f'F1 score: {f1_l2:.4f}')
```
