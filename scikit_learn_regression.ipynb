```python
import pandas as pd

cubicdata = pd.read_csv('cubnoise.dat')

cubicdata.head()
```

```python
import matplotlib.pyplot as plt
plt.plot(cubicdata['x'],cubicdata['y'],'.')
print(cubicdata['x'].shape)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

lin_reg=LinearRegression()
lin_reg.fit(cubicdata['x'].array.reshape(-1,1),cubicdata['y'])
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
plt.plot(cubicdata['x'],cubicdata['y'],'.')
x=np.linspace(-2,3,100)
y=lin_reg.coef_*x + lin_reg.intercept_
plt.plot(x,y,'r-')
```

```python
cubicdata.insert(0,'x^2',np.power(cubicdata['x'],2))
```

```python
cubicdata[['x^2','x']].values.shape
```

```python
pol=LinearRegression()
pol.fit(cubicdata[['x^2','x']].values,cubicdata['y'])
```

```python
print(pol.coef_)
print(pol.intercept_)
```

```python
plt.plot(cubicdata['x'],cubicdata['y'],'.')
y=pol.predict(x.reshape(-1,1))
plt.plot(x,y,'r-')
```

```python
cubicdata.insert(0,'x^3',np.power(cubicdata['x'],3))
```

```python
cubicdata.head()
```

```python
pol.fit(cubicdata[['x^3','x^2','x']].values,cubicdata['y'])
```

```python
print(pol.coef_)
print(pol.intercept_)
```

```python
plt.plot(cubicdata['x'],cubicdata['y'],'.')
y=pol.coef_[0]*x*x*x + pol.coef_[1]*x*x + pol.coef_[2]*x + pol.intercept_
plt.plot(x,y,'r-')
```
